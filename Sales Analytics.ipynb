### Data Exploration and Visualization

```python
# Summary of the data
df.describe()
```

```python
#Q1. Print out, then visualize the most popular payment method.
df.groupby('Payment Method')['Payment Method'].count().sort_values(ascending=False)
# .head(1) for the most popular
```

```python
df.groupby('Payment Method')['Payment Method'].count().sort_values(ascending=False).head(1)
```

```python
#Q2. Print out, then visualize the most popular item purchased.
base_color = sb.color_palette()[0]

method = df['Payment Method'].value_counts().index
# method
sb.countplot(data = df, y = 'Payment Method', color = base_color, order = method)
```

```python
# Q3. Print out the most popular item purchased by females
df[(df['Gender']=='Female')].groupby('Item Purchased')['Item Purchased'].count().sort_values(ascending=False).head(1)
```

```python
#Q4. Visualize the locations that drive majority of the sales
location = df['Location'].value_counts().head(10).index
# method
sb.countplot(data = df, y = 'Location', color = base_color, order = location)
```

```python
#Q5. On average how much do customers spend

mean_purchases = round(df['Purchase Amount (USD)'].mean(),2)
median_purchases = round(df['Purchase Amount (USD)'].median(),2)

print('Mean = '+str(mean_purchases))
print('Median = '+str(median_purchases))
```
