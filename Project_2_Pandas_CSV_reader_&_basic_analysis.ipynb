### **Adding or Removing Columns**

```python
coffee['price'] = 4.99
```

```python
coffee.head()
```

```python
import numpy as np
coffee['new_price'] = np.where(coffee['Coffee Type']=='Espresso', 3.99, 5.99)
```

```python
coffee.head()
```

```python
coffee.drop(0)
```

```python
coffee
```

```python
coffee.drop(columns = ["price"])
```

```python
coffee #see! it's not modified
```

```python
#If we want to actually modify
coffee.drop(columns = ["price"], inplace = True)
```

```python
coffee #see! it got modified
```

```python
coffee['revenue'] = coffee["Units Sold"] * coffee["new_price"]
```

```python
coffee
```

```python
coffee.rename(columns={'new_price':'price'})
```

```python
coffee.groupby(['Coffee Type'])['Units Sold'].sum()
```
